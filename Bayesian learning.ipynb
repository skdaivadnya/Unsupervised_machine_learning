```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, centers=3, cluster_std=1, n_features=2,random_state=0)
print(X.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1], c=y)
```

# aglomerative clustering 

```python
from sklearn.cluster import AgglomerativeClustering
aggclu = AgglomerativeClustering(n_clusters = 3)
cfit_ = aggclu.fit(X)
cfit_.labels_
```

```python
import numpy as np
np.unique(cfit_.labels_, return_counts=True)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=cfit_.labels_)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```python
Z = linkage(X, 'ward')
fig = plt.figure(figsize=(15, 15))
dn = dendrogram(Z)
```

```python
Z = linkage(X, 'single')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z)
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
kmeans = KMeans(n_clusters=2, random_state=0, n_init=1).fit(X)
kmeans.labels_
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
```

```python
kmeans.inertia_
```

```python
inertias = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init=1).fit(X)
    inertias.append(kmeans.inertia_)
```

```python
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
```

```python

plt.plot(range(1,10), inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
```
